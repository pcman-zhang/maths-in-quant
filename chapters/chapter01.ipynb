# 第1章 量化金融的数学地图

> **本章目标**: 在正式攀登之前, 先站在山顶俯瞰全貌. 理解量化金融的核心命题是什么, 一个专业的量化流程包含哪些步骤, 每个步骤在量化什么、用什么数学、看什么指标. 本章不引入复杂公式, 只建立直觉和方向感.

## 1.1 为什么需要一张"地图"

想象一下，你走进一座巨大的图书馆，书架上摆满了各种数学教材：微积分、概率论、线性代数、随机过程、偏微分方程、凸优化、机器学习……每一本都厚得像砖头。你的目标是"学会量化金融"，但你该从哪一本开始？哪些章节是必读的？哪些可以先跳过？这些知识之间又是如何相互关联的？

**没有地图的学习，是盲目的。**

很多自学者在量化路上半途而废，不是因为不够聪明，而是因为迷失在数学的丛林中——今天看了一点微积分，明天翻了几页概率论，后天被随机分析里的希腊字母吓退。学了一年，发现自己仍然说不清"做量化到底需要哪些数学"。

本章就是要为你绘制这张**数学知识地图**。它不是详细的教科书，而是导航图。当你读完后续章节再回头看这张图时，你会恍然大悟："原来第21章的GARCH模型，是建立在第7章概率论和第19章时间序列基础之上的；原来第25章的Black-Scholes公式，需要第3章的导数、第9章的期望和第24章的伊藤引理共同支撑。"

## 1.2 量化金融的核心命题: 在不确定性下做决策

在进入流派和工具之前, 我们必须先回答一个更根本的问题: **量化金融到底在解决什么问题?**

### 1.2.1 唯一的元问题

金融市场的本质特征可以用一句话概括: **未来不可知, 但决策必须做.** 你今天就要决定买什么、买多少, 但你不知道明天的价格. 你今年就要配置资产, 但你不知道明年的经济是扩张还是衰退. 你此刻就要给期权报价, 但你不知道到期时的标的价格.

量化金融——无论哪个流派——都是在回答同一个元问题:

> **如何在不确定的未来面前, 做出当下最优的资本配置决策?**

这个问题的每一个词都经过了精心的选择:

- **不确定的未来**: 这是概率论的地盘. 我们不说"明天涨 5%", 我们说"明天收益率服从某个分布". 不确定性不是需要消除的噪音——它是建模的对象本身.
- **当下最优**: 这是优化理论的地盘. "最优"需要一个目标函数——是要最大化收益? 最小化风险? 还是在两者之间找到一个平衡? 不同的目标会通向完全不同的策略.
- **资本配置决策**: 这是金融学的根本. 资本是稀缺的. 每一分钱分配给 A 资产, 就不能同时分配给 B 资产. 配置决策必然涉及**权衡(Trade-off)**.

### 1.2.2 核心权衡: 收益 vs 风险

这个元问题的内部, 隐藏着一个根本性的张力——**收益和风险是一枚硬币的两面**. 你无法只要其中一面:

- 存银行: 几乎零风险, 但收益也几乎为零(甚至跑不赢通胀).
- 买股票: 长期收益高, 但短期波动大, 可能亏钱.
- 卖期权: 大概率赚权利金(小钱), 小概率亏大钱(黑天鹅).

量化金融的全部数学, 本质上都是在**量化这个权衡**: 用期望度量收益, 用方差(或 VaR/CVaR)度量风险, 用夏普比率(或信息比率)度量"单位风险换来的收益". 你在本书中学到的每一个数学工具, 最终都将服务于这个权衡:

| 你想知道... | 你需要... | 对应章节 |
|-----------|---------|---------|
| 这个策略平均能赚多少? | 期望(一阶矩) | 第9章 |
| 这个策略的收益有多不稳定? | 方差/波动率(二阶矩) | 第9章 |
| 两只股票是一起涨还是跷跷板? | 协方差/相关系数 | 第9章 |
| 极端亏损的可能性有多大? | 分位数/VaR/CVaR | 第4, 8章 |
| 收益是否显著不等于零(还是运气)? | 假设检验 | 第11章 |
| 如何分配仓位才能最大化夏普? | 组合优化 | 第17-18章 |
| 期权在这种不确定性下值多少钱? | 随机分析+MC模拟 | 第24-26章 |
| 市场状态变了, 我的策略还灵吗? | 贝叶斯更新 | 第7章 |

### 1.2.3 数学和金融界为这个元问题做了什么: 四次范式跃迁

从 1952 年至今, 量化金融经历了四次关键的思想跃迁——每一次都用一个更深刻的数学框架重新回答了"如何在不确定性下做决策".

**第一次跃迁: Markowitz (1952) —— 从"选股"到"选组合"**

在 Markowitz 之前, 投资就是"挑好股票". 他的洞见是: **重要的不是单只股票的好坏, 而是它们组合在一起后的表现.** 两只各自有风险的股票, 如果它们不完全同涨同跌, 组合在一起的风险会低于各自风险的加权平均——这就是分散化的数学本质.

他用一个二次优化问题重新定义了投资:

$$\min_{w} \; w^T \Sigma w \quad \text{s.t.} \quad w^T \mu = \mu_{\text{target}},\; \sum w_i = 1$$

这看似简单的一行, 把"投资"从艺术变成了数学.

**第二次跃迁: Sharpe (1964) —— 从"总风险"到"系统性风险"**

Sharpe 发现了一个关键事实: 你无法通过分散化消除所有风险. 风险可以分解为两部分:

$$\sigma_i^2 = \underbrace{\beta_i^2 \sigma_m^2}_{\text{系统性风险(不可分散)}} + \underbrace{\sigma_{\epsilon_i}^2}_{\text{特质风险(可分散)}}$$

这意味着: 市场只会为**不可分散的风险**付你钱. 你承担了可以被分散掉的风险(特质风险), 市场不会给你任何补偿. CAPM 从这个洞察中诞生, 而 $\beta$ 从此成为量化金融中使用频率最高的希腊字母之一.

**第三次跃迁: Black-Scholes-Merton (1973) —— 从"预测价格"到"复制收益"**

衍生品定价曾被认为是不可能的问题: 期权的价值取决于未来股价, 而未来股价未知. Black, Scholes 和 Merton 的革命性洞见是: **你不需要预测股价. 只要你能用股票和债券动态复制出期权的收益结构, 期权的价格就必须等于复制成本——否则就存在无风险套利.**

这个思想的数学表达是一道偏微分方程:

$$\frac{\partial V}{\partial t} + \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} + rS\frac{\partial V}{\partial S} - rV = 0$$

它后来为 Scholes 和 Merton 赢得了 1997 年的诺贝尔经济学奖.

**第四次跃迁: 因子模型与统计学习 (1980s-至今) —— 从"单因子"到"高维预测"**

Fama-French(1993)将 Sharpe 的单因子拓展为三因子; 此后学术界和业界不断发现新的因子(动量, 质量, 低波动...). 因子模型的核心思想是:

$$R_i = \alpha_i + \beta_{i1}F_1 + \beta_{i2}F_2 + \cdots + \beta_{iK}F_K + \epsilon_i$$

它的量化含义是: **寻找 $\alpha_i > 0$ 且统计显著的股票**, 即那些"因子模型无法解释的超额收益". 这直接催生了现代对冲基金的核心业务: 做多 $\alpha$ 为正的资产, 做空 $\alpha$ 为负的资产, 对冲掉所有 $\beta$ 暴露, 赚取纯粹的 $\alpha$.

Grinold & Kahn(1999)用一个优雅的公式总结了这一范式的数学本质:

$$\text{IR} \approx \text{IC} \times \sqrt{\text{Breadth}}$$

信息比率 = 预测能力 × 预测广度. **量化策略不要求你比主动基金经理更聪明(高 IC), 只要求你做更多次独立预测(高 Breadth).** 这就是量化投资的哲学基石.

### 1.2.4 你将在本书中学到什么

总结起来, 量化金融的数学大厦建立在这四次跃迁之上. 你在本书中学习的每一个概念, 都能在这四次跃迁中找到自己的位置:

- Markowitz → 期望/方差/协方差(第9章) + 组合优化(第17-18章)
- Sharpe → 回归分析(第12章) + 系统性风险分解(第9章)
- Black-Scholes → 随机微积分(第24章) + 蒙特卡洛(第26章)
- 因子模型 → 假设检验(第11章) + 线性代数(第13-15章) + 机器学习(第27-29章)

**这不是四座孤立的岛屿——它们共享着概率统计这个共同的海洋, 由优化理论架起彼此之间的桥梁.**

### 1.2.5 专业量化流程全景: 七个步骤, 一张表

理论讲完了, 让我们落地. 在一个专业量化机构中, 策略从研究到实盘通常经历七个步骤. 下表将每一步的核心问题、数学工具和关键指标浓缩在一起:

| 步骤 | 在量化什么? | 核心数学工具(本书对应章节) | 关键量化指标 |
|------|-----------|-------------------------|------------|
| **1. 数据清洗** | 数据的完整性与准确性——垃圾进, 垃圾出 | 描述性统计, 缺失值插补, Winsorization(缩尾) | 缺失率, 异常值比例(Z>4), 标的覆盖率 |
| **2. 因子研究** | 信号对未来收益的预测能力 | 回归分析(Ch12), 信息系数(IC), 分层回测 | Rank IC>0.03, ICIR>0.5, Q1→Q5 收益单调性 |
| **3. 风险模型** | 资产联动关系与组合总风险暴露 | 协方差矩阵估计(Ch9), 收缩估计, PCA(Ch28) | 年化波动率<20-30%, 因子β暴露, RC%, 跟踪误差<5-8% |
| **4. 组合优化** | 信号→持仓权重的翻译, 收益/风险/成本的最优平衡 | 二次规划/凸优化(Ch17-18), 拉格朗日乘子 | 夏普比率>1, max drawdown<30%, 月换手率<50% |
| **5. 回测验证** | 历史表现是否为"真 alpha"而非过拟合噪音 | 样本外检验(Ch11), Bootstrap(Ch10), PBO | OOS Sharpe>0, Calmar 比率, PBO<0.1 |
| **6. 执行交易** | 纸面信号→真实成交的摩擦成本 | Almgren-Chriss 冲击模型, VWAP 算法 | 滑点(bps), Implementation Shortfall |
| **7. 归因监控** | 盈亏来源分解 + 策略衰减检测 | Brinson 归因, 因子归因(Ch12), 贝叶斯更新(Ch7) | Alpha>0, 因子贡献分解, 滚动 Sharpe 趋势 |

这七个步骤**不是一次性的流水线, 而是一个持续运行的闭环**: 归因(7)的结果反馈回因子研究(2)和风险模型(3)——如果策略跑输了, 是因子失效了? 还是风险约束太紧了? 本书的章节编排正是按照这个流程设计的: 第 1-6 章给数据处理语言, 第 7-12 章给统计推断工具, 第 13-15 章给矩阵运算, 第 16-18 章给组合优化解法, 第 19-22 章给时间序列视角, 第 23-26 章给随机分析, 第 27-29 章给机器学习的现代武器.

本书并没有系统地讲解**数据清洗**部分. 原因有三: 第一, 我不希望本书沦为 Python 编程工具书——数据清洗和质量探查是基本的 Python 语言能力, 不赘述. 第二, 目前市面上免费或积分制社区数据源的数据质量参差不齐, 不适合作为严肃量化投研的基础. 第三, 本书自身使用的 `ifind_price_data.csv` 也只是一个**教学样本**(约 3 年, 3 只标的)——它足够支撑你理解数学概念, 但远不足以支撑实盘策略开发.

如果你希望认真地做量化投研, 你需要的是**持牌的金融数据提供商**. 持牌机构的数据不仅质量上乘, 更重要的是能提供大量的**另类数据**(分析师预期, 供应链图谱, 新闻情绪, 卫星图像等)——这些数据已经是专业量化机构 Alpha 因子的重要来源. 数据是量化策略的原材料, 原材料的质量决定了成品的天花板.

### 1.2.6 一个重要的免责声明

需要特别说明的是: **本教程是一本数学教材, 而非量化交易系统教材.** 我不会教你如何搭建一套高频交易引擎, 不会讨论最新的 GPU 加速回测架构, 不会深入订单簿的微观结构, 也不会告诉你哪个因子"现在还能用". 这些都是快速变化、高度工程化、与机构具体基础设施深度绑定的领域——它们不适合写成一本面向高中数学基础的通用教材.

**本教程唯一的目标是: 让你能够从数学的角度理解量化过程.** 当你读完本书后, 看到一份量化策略的研报或风险报告, 你不会被里面的公式吓退——你会知道那个 $\Sigma$ 是协方差矩阵, 知道 $\beta$ 是对市场风险的暴露, 知道 IC 是衡量预测能力的指标, 知道那个优化问题在求解什么. 你不会成为一个"调参侠"——你会成为一个**理解自己在做什么的量化实践者**.

数学是量化大厦的地基. 工程工具会过时, 因子会失效, 市场结构会变化——但数学不会. 方差永远是 $E[(X-\mu)^2]$, 贝叶斯公式永远是 $P(H|D) \propto P(D|H)P(H)$, 无论你用的是 2024 年的 PyTorch 还是 2034 年的量子计算机.

## 1.3 量化金融的四大流派

量化金融不是一门单一的学科, 它包含了多个研究方向, 每个方向对数学的"口味"各不相同. 我们可以将其大致划分为四大流派:

### 1.3.1 资产配置（Asset Allocation）

**核心问题**：手上有100元，该买多少股票、多少债券、多少黄金？

这是量化最古老也最经典的问题。1952年，Harry Markowitz 用一篇论文开创了现代投资组合理论。他的核心思想是：**投资不是在寻找"最好的单一资产"，而是在寻找"最好的资产组合"**。通过分散投资，可以在不降低预期收益的情况下减少风险（或者用金融术语说——在相同风险下提高收益）。

**数学口味**：微积分（理解收益率和风险的度量）、概率统计（描述资产收益的不确定性）、线性代数（处理多资产的协方差矩阵）、优化理论（求解最优权重）。

**代表模型**：均值-方差模型（Mean-Variance）、Black-Litterman 模型、风险平价（Risk Parity）。

### 1.3.2 因子投资（Factor Investing）

**核心问题**：什么样的股票未来更可能涨？

如果说资产配置解决的是"买多少"，因子投资解决的就是"买什么"。因子投资者相信，股票的收益可以被少数几个系统性因素解释——比如价值因子（低市盈率股票跑赢高市盈率）、动量因子（过去涨的股票继续涨）、质量因子（盈利稳定的公司更可靠）。

**数学口味**：回归分析（检验因子与收益的关系）、假设检验（判断因子是否"真的"有效）、线性代数（多因子模型的矩阵表达）。

**代表模型**：CAPM（资本资产定价模型）、Fama-French 三因子/五因子模型、Barra 多因子风险模型。

### 1.3.3 衍生品定价（Derivatives Pricing）

**核心问题**：一个期权应该值多少钱？

衍生品（期权、期货、互换）的价值依赖于底层资产（如股票、利率、大宗商品）的价格。与股票投资"预测涨跌"不同，衍生品定价的核心是**在无套利假设下，用数学推导出"公平价格"**。这个领域可能是量化金融中数学最"硬核"的分支。

**数学口味**：随机分析（布朗运动、伊藤引理）、偏微分方程（BS PDE）、数值方法（蒙特卡洛模拟、有限差分）。

**代表模型**：Black-Scholes-Merton 期权定价模型、二叉树模型、局部波动率与随机波动率模型。

### 1.3.4 统计套利与高频交易（StatArb & HFT）

**核心问题**：市场上有哪些可重复的、短期的定价偏差可以利用？

统计套利者不赌"大盘涨跌"，而是寻找**成对或多组资产之间的相对价格关系**。当这种关系暂时偏离历史常态时，就做多被低估的、做空被高估的，等待关系回归。高频交易则将这种思想推向了毫秒甚至微秒级别。

**数学口味**：时间序列分析（平稳性、协整）、概率统计（厚尾分布、极值理论）、优化理论（执行成本最小化）、机器学习（模式识别）。

**代表模型**：配对交易（Pairs Trading）、均值回归策略、市场微观结构模型。

## 1.4 数学知识的分层结构

如果把量化金融比作一座大厦，数学就是它的地基和钢架。这座大厦的地基不是一块平地，而是分层的：

### 第一层：数学基础（第1-6章）

这是你高中就已经接触过的工具，但我们要把它们"打磨"成金融建模的利器：

| 数学工具 | 在量化中的作用 |
|---------|--------------|
| **函数与极限** | 描述价格与收益率之间的非线性关系；理解连续复利的极限过程 |
| **导数与微分** | 度量敏感性：债券久期、期权 Greeks（Delta, Gamma, Vega）本质上都是导数 |
| **积分与累积** | 计算预期收益、现金流贴现、概率加权平均值 |
| **对数与指数** | 对数收益率（时间可加性）；连续复利公式；几何平均收益 |
| **级数与泰勒展开** | 对复杂函数做局部线性/二次近似；BS 公式的数值逼近 |

> 💡 **关键直觉**：在量化中，导数不是"斜率"而是"敏感度"，积分不是"面积"而是"累积"，对数不是"计算工具"而是"收益率的自然语言"。

### 第二层：概率统计与线性代数（第7-15章）

这是量化的**核心语言**。金融数据本质上是随机的，必须用概率来描述：

| 数学工具 | 在量化中的作用 |
|---------|--------------|
| **概率论基础** | 描述市场的不确定性；贝叶斯思维用于策略更新 |
| **随机变量与分布** | 收益率服从什么分布？正态？t分布？对数正态？ |
| **期望、方差、协方差** | 收益 = 期望；风险 = 方差；分散化 = 协方差 |
| **大数定律与中心极限定理** | 为什么历史回测有意义；为什么均值趋向正态 |
| **假设检验与置信区间** | 判断策略收益是"真有效"还是"运气" |
| **回归分析** | 因子投资的核心工具；Alpha/Beta 的数学来源 |
| **线性代数** | 多资产组合的矩阵表达；协方差矩阵的特征分解 |

> 💡 **关键直觉**：期望和方差不是抽象的数学概念，它们就是金融语言里的"收益"和"风险"。当你看到 $E[R]$，读作"期望收益"；看到 $\sigma^2$，读作"风险"。

### 第三层：进阶方法（第16-26章）

当你掌握了基础工具后，就可以用它们构建更复杂的模型：

| 数学工具 | 在量化中的作用 |
|---------|--------------|
| **优化理论** | Markowitz 组合优化；风险预算；交易成本最小化 |
| **时间序列分析** | ARIMA 预测收益；GARCH 预测波动率；协整检验 |
| **随机分析** | 衍生品定价的数学基础；伊藤引理是随机世界的"链式法则" |
| **蒙特卡洛模拟** | 当公式解不出来时，用计算机"模拟"出答案 |

### 第四层：机器学习（第27-29章）

将经典统计工具扩展到高维、非线性的场景：

| 数学工具 | 在量化中的作用 |
|---------|--------------|
| **监督学习** | 用逻辑回归、随机森林预测涨跌方向 |
| **PCA 降维** | 从数百个相关因子中提取独立风险源 |
| **聚类分析** | 市场状态识别；股票行业分类 |

## 1.5 各流派与数学的对应关系

为了更直观地理解"哪个流派需要哪些数学"，我们来看下面这张对应表：

| 数学分支 | 资产配置 | 因子投资 | 衍生品定价 | 统计套利/HFT |
|---------|:-------:|:-------:|:---------:|:-----------:|
| 微积分（导数/积分） | ⭐⭐⭐ | ⭐⭐ | ⭐⭐⭐⭐⭐ | ⭐⭐ |
| 概率统计 | ⭐⭐⭐⭐ | ⭐⭐⭐⭐⭐ | ⭐⭐⭐⭐⭐ | ⭐⭐⭐⭐⭐ |
| 线性代数 | ⭐⭐⭐⭐⭐ | ⭐⭐⭐⭐ | ⭐⭐⭐ | ⭐⭐⭐ |
| 优化理论 | ⭐⭐⭐⭐⭐ | ⭐⭐⭐ | ⭐⭐⭐ | ⭐⭐⭐⭐ |
| 时间序列 | ⭐ | ⭐⭐ | ⭐⭐ | ⭐⭐⭐⭐⭐ |
| 随机分析 | ⭐ | ⭐ | ⭐⭐⭐⭐⭐ | ⭐⭐ |
| 机器学习 | ⭐⭐ | ⭐⭐⭐⭐ | ⭐⭐ | ⭐⭐⭐⭐ |

> ⭐ 越多，表示该数学分支在该流派中越重要。

**观察这张表，你会发现几个有趣的规律：**

1. **概率统计是所有流派的通用语言**——无论做哪个方向，你都必须学好概率统计。
2. **线性代数是资产配置的"命门"**——多资产组合的协方差矩阵、优化求解，离开了矩阵运算寸步难行。
3. **随机分析是衍生品定价的"入场券"**——不懂布朗运动和伊藤引理，就无法理解 Black-Scholes 公式从何而来。
4. **时间序列是统计套利的"看家本领"**——配对交易、均值回归策略，本质上都是在建模时间序列的依赖关系。

## 1.6 本教程的攀登路径

本教程将沿着一条精心设计的路线，带你从高中数学一步步攀登到量化前沿。这条路线遵循四个原则：

### 原则一：先单后多
先讲清楚"单一资产的收益和风险怎么算"（第3-5, 7-10章），再扩展到"多个资产如何组合"（第13-17章）。你只有先理解了单个随机变量的期望和方差，才能理解为什么两个随机变量的协方差能让组合风险降低。

### 原则二：先静后动
先讲"某一时刻的截面数据"（如某天所有股票的市盈率），再讲"随时间演变的时间序列"（如某只股票过去一年的价格走势）。截面分析是时间序列分析的基础。

### 原则三：先频率后随机
先讲"确定性的优化"（在给定预期收益下最小化风险，第16-18章），再讲"随机过程的世界"（价格不是确定性的函数，而是随机游走，第23-26章）。确定性优化为理解随机优化提供了直觉基础。

### 原则四：先解析后数值
每个阶段都是先给出"解析解"（能用公式写出来的精确答案），再给出"数值方法"（当公式解不出来时，用计算机逼近答案）。这培养了你"能推公式、也能写代码"的双重能力。

## 1.7 Python 示例：绘制量化金融数学知识图谱

本章的核心代码任务，是用 Python 将上述的数学依赖关系可视化。我们使用 Matplotlib 绘制一张分层的知识图谱，展示从基础数学到量化应用的完整路径。

```
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['WenQuanYi Micro Hei'] 
plt.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots(figsize=(18, 14))
ax.set_xlim(0, 18)
ax.set_ylim(0, 14)
ax.axis('off')

# ============ 定义四层节点坐标 ============
# 基础层（第1-6章）
math_basic = {
    '高中数学': (3, 1.5),
    '函数与极限': (6, 1.5),
    '导数与积分': (9, 1.5),
    '对数与级数': (12, 1.5),
    '级数展开': (15, 1.5),
}

# 核心层（第7-15章）
math_core = {
    '概率论': (3, 4.5),
    '随机变量与分布': (6, 4.5),
    '期望方差协方差': (9, 4.5),
    '线性代数': (12, 4.5),
    '矩阵运算特征分解': (15, 4.5),
}

# 进阶层（第16-26章）
math_advanced = {
    '假设检验回归分析': (3, 7.5),
    '时间序列ARIMA/GARCH': (6, 7.5),
    '优化理论凸优化': (9, 7.5),
    '随机分析伊藤引理': (12, 7.5),
    '蒙特卡洛模拟': (15, 7.5),
}

# 应用层（量化实践）
applications = {
    '资产配置均值-方差': (2.5, 11),
    '因子投资CAPM/FF': (5.5, 11),
    '统计套利配对交易': (8.5, 11),
    '衍生品定价BS模型': (11.5, 11),
    '风险管理VaR/CVaR': (14.5, 11),
}

# ============ 定义连接关系 ============
connections = [
    # 基础 -> 核心
    ('高中数学', '概率论'), ('高中数学', '线性代数'),
    ('函数与极限', '随机变量与分布'),
    ('导数与积分', '期望方差协方差'), ('导数与积分', '优化理论凸优化'),
    ('对数与级数', '假设检验回归分析'),
    ('级数展开', '蒙特卡洛模拟'),
    # 核心 -> 进阶
    ('概率论', '假设检验回归分析'), ('概率论', '时间序列ARIMA/GARCH'),
    ('随机变量与分布', '随机分析伊藤引理'),
    ('期望方差协方差', '优化理论凸优化'),
    ('线性代数', '优化理论凸优化'),
    ('矩阵运算特征分解', '蒙特卡洛模拟'),
    # 进阶 -> 应用
    ('假设检验回归分析', '因子投资CAPM/FF'),
    ('假设检验回归分析', '统计套利配对交易'),
    ('时间序列ARIMA/GARCH', '统计套利配对交易'),
    ('时间序列ARIMA/GARCH', '风险管理VaR/CVaR'),
    ('优化理论凸优化', '资产配置均值-方差'),
    ('优化理论凸优化', '风险管理VaR/CVaR'),
    ('随机分析伊藤引理', '衍生品定价BS模型'),
    ('蒙特卡洛模拟', '衍生品定价BS模型'),
    ('蒙特卡洛模拟', '风险管理VaR/CVaR'),
]

# ============ 绘制连接线 ============
all_nodes = {**math_basic, **math_core, **math_advanced, **applications}
for start, end in connections:
    ax.annotate('', xy=all_nodes[end], xytext=all_nodes[start],
               arrowprops=dict(arrowstyle='->', color='#999999', lw=1.2, alpha=0.5))

# ============ 绘制节点 ============
def draw_nodes(node_dict, facecolor, edgecolor, ax):
    for name, (x, y) in node_dict.items():
        box = mpatches.FancyBboxPatch(
            (x-1.1, y-0.5), 2.2, 1.0,
            boxstyle="round,pad=0.05,rounding_size=0.2",
            facecolor=facecolor, edgecolor=edgecolor,
            linewidth=2.5, alpha=0.9
        )
        ax.add_patch(box)
        ax.text(x, y, name, ha='center', va='center', 
                fontsize=10, fontweight='bold', color='#333333')

# 按层次绘制，颜色由浅入深
draw_nodes(math_basic, '#E8F5E9', '#4CAF50', ax)      # 浅绿
draw_nodes(math_core, '#BBDEFB', '#2196F3', ax)       # 浅蓝
draw_nodes(math_advanced, '#FFE0B2', '#FF9800', ax)   # 浅橙
draw_nodes(applications, '#F8BBD0', '#E91E63', ax)    # 浅粉

# ============ 添加层次标签 ============
layer_labels = [
    (0.5, 1.5, '基础层(第1-6章)', '#4CAF50'),
    (0.5, 4.5, '核心层(第7-15章)', '#2196F3'),
    (0.5, 7.5, '进阶层(第16-26章)', '#FF9800'),
    (0.5, 11, '应用层(量化实践)', '#E91E63'),
]
for x, y, label, color in layer_labels:
    ax.text(x, y, label, ha='center', va='center', fontsize=11,
            fontweight='bold', color=color, rotation=90)

# ============ 标题与图例 ============
ax.text(9, 13.2, '量化金融数学知识图谱', ha='center', va='center',
        fontsize=20, fontweight='bold', color='#1a1a1a')
ax.text(9, 12.6, 'Quantitative Finance Mathematical Knowledge Graph',
        ha='center', va='center', fontsize=12, color='#666666')

legend_elements = [
    mpatches.Patch(facecolor='#E8F5E9', edgecolor='#4CAF50', label='数学基础（第1-6章）'),
    mpatches.Patch(facecolor='#BBDEFB', edgecolor='#2196F3', label='概率统计与线性代数（第7-15章）'),
    mpatches.Patch(facecolor='#FFE0B2', edgecolor='#FF9800', label='进阶方法（第16-26章）'),
    mpatches.Patch(facecolor='#F8BBD0', edgecolor='#E91E63', label='量化应用'),
]
ax.legend(handles=legend_elements, loc='lower center', bbox_to_anchor=(0.5, -0.02),
         ncol=4, fontsize=10, frameon=True, fancybox=True)

plt.tight_layout()
plt.show()
```

运行上述代码，你将得到一张完整的知识依赖图（如本章附图所示）。这张图的价值在于：

- **横向看**：同一层内的不同节点，代表了同一抽象层次上的不同工具。
